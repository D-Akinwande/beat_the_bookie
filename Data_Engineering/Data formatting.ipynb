```python
### Remove empty row
Clean_Data = Data.dropna()

print(Data.iloc[5700,:])
print()
print(Clean_Data.iloc[5700,:])
```

```python
### Get referee column
y = Clean_Data.loc[:,'Referee']

y
```

```python
### Gets different referee names
cat_y = y.unique()

print(cat_y)
print()
```

```python
### Indices 0-29 (inclusive) are full names, e.g. Rob Harris
### Indices 30-69 are initials, e.g. N. S. Barry
### Indices 70-96 are last name first initials, e.g. Wiley, A. G. and Yates, N
### Indices 97-179 are no dot initials, e.g. D Elleray and †U Rennie

index = 178
print(cat_y[index])
print()
```

```python
### Puts different types of data into different categories
cat_y_1 = cat_y[0:29]
cat_y_2 = cat_y[30:69]
cat_y_3 = cat_y[70:96]
cat_y_4 = cat_y[97:]

print(cat_y_1)
print()
print(cat_y_2)
print()
print(cat_y_3)
print()
print(cat_y_4)
print()
```

```python
### Removes the duplicates with † in the 4th category
j = 0

for i in cat_y_4:
    if i.find('†') == 0:
        cat_y_4 = np.delete(cat_y_4,j,axis = None)
        j -= 1
    j += 1    

print(cat_y_4)
print()
```

```python
### Gets last names of 4th category
def getLastNames_1_4(cat_y_4):
    cat_y_4_surname = np.array([])

    for i in cat_y_4:
        n = i.find(' ')
        lastname = i[n+1:]
        #print(lastname)
        cat_y_4_surname = np.append(cat_y_4_surname,np.array([lastname]))

    return cat_y_4_surname

cat_y_4_surname = getLastNames_1_4(cat_y_4)
print(cat_y_4_surname)
print()
```

```python
### Finds people with the same initials and surname and removes them
cat_y_4_2 = np.asarray(cat_y_4, dtype=str)
repeats = np.array([])
k = 0
m = 0

for i in cat_y_4_surname:
    num_names = np.char.count(cat_y_4_2, i)
    
    if np.sum(num_names) > 1:
        #print(i)
        #print(num_names)
        #print(cat_y_4_2[num_names == 1])
        
        dups = cat_y_4_2[num_names == 1]
        #print(dups[0][0].capitalize())
        #print(dups[1][0].capitalize())
        
        if dups[0][0].capitalize() == dups[1][0].capitalize():
            cat_y_4_2 = np.delete(cat_y_4_2,k,axis = None)
            k -= 1
        else:
            repeats = np.append(repeats,dups)
    k += 1
    m += 1

### Formats results and adds missed repeated names from below
cat_y_4_2 = np.append(cat_y_4_2,'P Jones')
cat_y_4_surname = np.unique(cat_y_4_surname)
repeats = np.unique(repeats)
repeats = np.append(repeats,'P Jones')
repeats_surnames = np.unique(getLastNames_1_4(repeats)) 

print()
print(cat_y_4_2)
print()
print(cat_y_4_surname)
print()
print(repeats)
print()
print(repeats_surnames)
print()
```

```python
### Formats unique referee name column (to find missed repeated names and as a test for function)

def formatRefNames(cat_y,cat_y_4_surname,cat_y_4_2):
    cat_y_copy = np.asarray(cat_y,dtype = str)

    for i in cat_y_4_surname:
        num_names_1 = np.char.count(cat_y_copy, i)
        num_name_2 = np.char.count(cat_y_4_2, i)
        full_name = cat_y_4_2[num_name_2 == 1]
        #print(full_name)

        if np.shape(full_name)[0] > 1:
            #print(full_name)
            rng = np.random.default_rng()
            names = cat_y_copy[num_names_1 == 1]
            full_name_2 = np.array([])
            found = False
            #print(names)

            for j in range(0,len(names)):
                
                ### Checks to see if last name is first
                if names[j].find(i) == 0:

                    for k in range(0,len(full_name)):

                        if names[j][len(i)+2] == full_name[k][0]:
                            full_name_2 = np.append(full_name_2,full_name[k]) 
                            found = True
                            break
                    
                    ### If no match is found, randomly choose from first two names
                    if found == False:
                        if rng.integers(2) == 0:
                            full_name_2 = np.append(full_name_2,full_name[0])
                        else:
                            full_name_2 = np.append(full_name_2,full_name[1])
                    #print(full_name_2)

                else:
                    for k in range(0,len(full_name)):

                        if names[j][0] == full_name[k][0]:
                            full_name_2 = np.append(full_name_2,full_name[k])
                            found = True
                            break

                    if found == False:
                        if rng.integers(2) == 0:
                            full_name_2 = np.append(full_name_2,full_name[0])
                        else:
                            full_name_2 = np.append(full_name_2,full_name[1])
                    #print(full_name_2)
                #print(full_name_2)
            cat_y_copy[num_names_1 == 1] = full_name_2
        else:
            cat_y_copy[num_names_1 == 1] = full_name

    return cat_y_copy

cat_y_copy = formatRefNames(cat_y,cat_y_4_surname,cat_y_4_2)
print()
print(cat_y)
print()
print(cat_y_copy)
print()
print(np.unique(cat_y_copy))
print()
```

```python
### Formats array version of referee column

y_copy = formatRefNames(y,cat_y_4_surname,cat_y_4_2)

print()
print(y)
print()
print(y_copy)
print()
print(np.unique(y_copy))
print()
```

```python
### Converts formatted referee column to pandas series

index = np.asarray(range(0,5700))
index_2 = np.asarray(range(5701,len(y)+1))
full_index = np.concatenate((index, index_2), axis=None)

y_form = pd.Series(data=y_copy,index=full_index,name='Referee')

y_form
```
